## Step 10: Complete 17.2.3 WHERE (3 queries)

### Step 10 1st Query: Select records where Copyright is Greater than 2016

```python
import pandas as pd
connection = sqlite3.connect('books.db')
pd.read_sql("""SELECT title, edition, copyright
               FROM titles
               WHERE copyright > '2016'""", connection)
```

### Step 10 2nd Query: Select records in titles table where title contains How to Program

```python
pd.read_sql("""SELECT title, edition, copyright
                FROM titles 
                WHERE title LIKE '%How to Program%'""", 
             connection)
```

### Step 10 3rd Query: Select records in authors table where first name is either Dan or Paul and Sort records by last name

```python
pd.read_sql("""SELECT first, last
                FROM authors 
                WHERE first LIKE '%Dan%' or first LIKE '%Paul%' ORDER BY last""", connection)
```

## Step 11: Complete 17.2.4 ORDER BY (4 queries)

### Step 11 2nd Query: Select records in title table and sort by copyright

### Step 11 4th Query: Select records in title table and sort by title

## Step 12 Complete 17.2.5 INNER JOIN (1 query)

```python
import pandas as pd
connection = sqlite3.connect('books.db')
pd.read_sql("""SELECT first, last, isbn
               FROM authors
               INNER JOIN author_ISBN
                   ON authors.id = author_ISBN.id
                ORDER BY last, first""", connection).head()
```
